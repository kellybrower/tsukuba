```python
#scatterplot size/cap
var = 'capacity'
data = pd.concat([df['size'], df[var]], axis = 1)
data.plot.scatter(x = var, y = 'size', ylim = (0, 200));
```

```python
#scatter size/fare
var2 = 'fare'
data = pd.concat([df['size'], df[var2]], axis = 1)
data.plot.scatter(x = var2, y = 'size', ylim = (0, 200));
```

```python
#scatter cap/fare
var3 = 'fare'
data = pd.concat([df['capacity'], df[var3]], axis = 1)
data.plot.scatter(x = var3, y = 'capacity', ylim = (0, 400));
```

```python
var4 = 'period'
data = pd.concat([df['fare'], df[var4]], axis = 1)
data.plot.scatter(x = var4, y = 'fare', ylim = (0, 12500));
```

```python
#boxplot period/capacity
var5 = 'period' 
data = pd.concat([df['capacity'], df[var5]], axis = 1)
f, ax = plt.subplots(figsize=(10,10))
fig = sns.boxplot(x=var5, y='capacity', data = data)
fig.axis(ymin=0, ymax = 500);
plt.xticks(rotation=45);
```

```python
#correlation matrix
corrmat = df.corr()
f, ax = plt.subplots(figsize = (5, 4))
sns.heatmap(corrmat, vmax=1, square=True);
```

```python
#size correlation matrix
k = 9 #number of variables for heatmpa
cols = corrmat.nlargest(k, 'size')['size'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f',
                annot_kws = {'size':10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#scatterplot
sns.set()
cols = ['size', 'fare', 'capacity', 'period']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```
